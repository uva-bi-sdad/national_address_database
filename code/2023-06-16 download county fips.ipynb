```python
import pandas as pd
import requests
```

```python
r = requests.get('https://transition.fcc.gov/oet/info/maps/census/fips/fips.txt')
r
```

```python
lines = str(r.content).split('\\n')
lines
```

```python
relevant_lines = lines[72:]
relevant_lines
```

```python
df = pd.DataFrame()
```

```python
fips = []
names = []

for line in relevant_lines:
    fips.append(line[:10].split()[0])
    names.append(' '.join(line[10:].split()))

df['fips'] = fips
df['names'] = names

df = df[df['names'].apply(lambda x: len(x)> 2)]
df
```

```python
df.to_csv('../data/fips_county.csv', index=False)
```
